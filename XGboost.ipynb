```python
num_cols=santa.columns[santa.dtypes!='object']

def histograms(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(12,30))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name)
    fig.tight_layout()
    plt.show()

histograms(santa, num_cols, 12, 3)
```

```python
santa=santa.drop(['pais_residencia','indresi',],axis=1)
```

```python
santa.dtypes
```

```python
for var in santa.columns[santa.dtypes=='object']:
    santa[var]=santa[var].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
for var in santa.columns[santa.dtypes=='category']:
    santa[var]=LabelEncoder().fit_transform(santa[var])
```

```python
products_col=['ind_cco_fin_ult1','ind_cder_fin_ult1', 'ind_cno_fin_ult1', 'ind_ctju_fin_ult1','ind_ctma_fin_ult1', 
        'ind_ctop_fin_ult1', 'ind_ctpp_fin_ult1','ind_dela_fin_ult1', 'ind_ecue_fin_ult1', 'ind_fond_fin_ult1',
        'ind_hip_fin_ult1', 'ind_plan_fin_ult1', 'ind_pres_fin_ult1','ind_reca_fin_ult1', 'ind_tjcr_fin_ult1', 
        'ind_valo_fin_ult1','ind_nomina_ult1', 'ind_nom_pens_ult1', 'ind_recibo_ult1']

for var in products_col:
    santa[var]=santa[var].astype('int64')
```

```python
santa.head()
```

```python
santa_user=santa.drop(products_col, axis=1)
```

```python
santa_pro=santa[products_col]
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
distortions = []
K = range(1,20)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(santa_pro)
    distortions.append(sum(np.min(cdist(santa_pro, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / santa_pro.shape[0])

plt.plot(K, distortions)
plt.xlabel('no. of cluster')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeans=KMeans(n_clusters=10)
kmeans.fit(santa_pro)
santa_user['product']=kmeans.labels_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(santa_user.drop(['product'], axis=1), santa_user['product'], 
                                                test_size=.3, random_state=123)
```

```python
x_train.shape, y_train.shape
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(x_train, y_train)
dtest = xgb.DMatrix(x_test, y_test)

param= {'objective':'multi:softmax','num_class':10, 'eta': 0.01,'max_depth': 8,
        'min_child_weight':11,'silent':1,'subsample':0.7,'colsample_bytree': 0.7,
        'n_estimators':300,'missing':-999,'seed':1337}
```

```python
cv_xgb = xgb.cv(params = param, dtrain = dtrain, num_boost_round = 3000, nfold = 5,
                metrics = ['merror'], early_stopping_rounds = 100)

cv_xgb.tail()
```

```python
xgb_model = xgb.train(param, dtrain, num_boost_round = 64)
pred = xgb_model.predict(dtest)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,pred)
```
