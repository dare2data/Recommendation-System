```python

drop_variable=['ncodpers','indrel_1mes','tipodom','fecha_dato','conyuemp','ind_empleado','ult_fec_cli_1t','ind_ahor_fin_ult1', 'ind_aval_fin_ult1','ind_deco_fin_ult1', 'ind_deme_fin_ult1','ind_viv_fin_ult1']
santa=data.drop(drop_variable, axis=1)
```

```python
santa['fecha_alta']=pd.to_datetime(santa['fecha_alta'])
santa.insert(0,'day',santa['fecha_alta'].dt.day)
santa.insert(1,'month',santa['fecha_alta'].dt.month)
santa.insert(2,'year',santa['fecha_alta'].dt.year)

santa=santa.drop(['fecha_alta'], axis=1)
```

```python
santa['renta'].fillna(santa['renta'].mean(), inplace=True)
```

```python
santa[santa.isnull().any(axis=1)]
```

```python
santa=santa.dropna(how='any') 
```

```python
santa=santa.reset_index(drop=True)
```

```python
santa['age']=santa['age'].astype('int64')
santa['antiguedad']=santa['antiguedad'].astype('float')
```

```python
santa.head()
```

```python
num_cols=santa.columns[santa.dtypes!='object']

def histograms(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(12,30))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name)
    fig.tight_layout()
    plt.show()

histograms(santa, num_cols, 12, 3)
```

```python
santa=santa.drop(['pais_residencia','indresi',],axis=1)
```

```python
santa.dtypes
```

```python
for var in santa.columns[santa.dtypes=='object']:
    santa[var]=santa[var].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
for var in santa.columns[santa.dtypes=='category']:
    santa[var]=LabelEncoder().fit_transform(santa[var])
```

```python
products_col=['ind_cco_fin_ult1','ind_cder_fin_ult1', 'ind_cno_fin_ult1', 'ind_ctju_fin_ult1','ind_ctma_fin_ult1', 
        'ind_ctop_fin_ult1', 'ind_ctpp_fin_ult1','ind_dela_fin_ult1', 'ind_ecue_fin_ult1', 'ind_fond_fin_ult1',
        'ind_hip_fin_ult1', 'ind_plan_fin_ult1', 'ind_pres_fin_ult1','ind_reca_fin_ult1', 'ind_tjcr_fin_ult1', 
        'ind_valo_fin_ult1','ind_nomina_ult1', 'ind_nom_pens_ult1', 'ind_recibo_ult1']

for var in products_col:
    santa[var]=santa[var].astype('int64')
```

```python
santa.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(santa.drop(products_col, axis=1), santa[products_col], 
                                                test_size=.3, random_state=123)
```

```python
x_train.shape, y_train.shape
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)

x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
```

```python
n_input_node= 17
n_nodes_hl1=500
n_nodes_hl2=500
n_nodes_hl3=500
n_output_node = 19

epochs =5
batch_size = 100
```

```python
model = Sequential([
 Dense(output_dim=n_nodes_hl1, input_dim=n_input_node, activation='sigmoid'),
 Dense(output_dim=n_nodes_hl2, input_dim=n_nodes_hl1, activation='sigmoid'),
 Dense(output_dim=n_nodes_hl3, input_dim=n_nodes_hl2, activation='sigmoid'),
    
 Dense(output_dim=n_output_node, input_dim=n_nodes_hl3, activation='softmax')
 ])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
trained_model = model.fit(x_train, y_train, nb_epoch=epochs, batch_size=batch_size, 
                             validation_data=(x_test, y_test))
```
